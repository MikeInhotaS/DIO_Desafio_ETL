```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
vinhos = pd.read_csv(url, sep=';')
vinhos.head(10)
```

```python
vinhos.describe()
```

```python
vinhos.info()
```

```python
plt.hist(vinhos['quality'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Qualidade')
plt.ylabel('Contagem')
plt.title('Distribuição da Qualidade dos Vinhos')
plt.grid(True)

plt.show()
```

```python
def avalia_qualidade(qualidade):
  avaliacao="n"
  if qualidade <= 5:
    avaliacao= "Baixa"
  elif qualidade <= 7:
    avaliacao= "Media"
  else:
    avaliacao= "Alta"
  return avaliacao
```

```python
vinhos['avaliacao'] = vinhos['quality'].apply(avalia_qualidade)
vinhos.head(3)
```

```python
plt.hist(vinhos['avaliacao'], bins=10, color='darkgreen', edgecolor='black')
plt.xlabel('Avaliação')
plt.ylabel('Contagem')
plt.title('Distribuição da Avaliação dos Vinhos')
plt.grid(True)

plt.show()
```

```python
vinhos.to_excel('vinhos_avaliados.xlsx', index=False)
```
